```python
import pandas as pd
from io import StringIO
import random as rnd
```

```python

df_cols = [f'x{i:02}' for i in range(17)]
df_cols.append('y')
print(len(df_cols))

print(df_cols)
```

```python
str_cols = {}
str_cols["x00"] = [f'my_cat{i:02}' for i in range(3)]
str_cols["x01"] = [f'this_cat{i:02}' for i in range(4)]
str_cols["x02"] = [f'some_cat{i:02}' for i in range(5)]
str_cols
```

```python
data_stats_json = '{"x03":{"count":25870.0,"min":0.0,"max":45.0},"x04":{"count":25870.0,"min":0.005,"max":0.02},"x05":{"count":25870.0,"min":0.005,"max":0.025},"x06":{"count":25870.0,"min":0.00018,"max":0.13337},"x07":{"count":25870.0,"min":0.0025,"max":0.015},"x08":{"count":25870.0,"min":2704.0,"max":16981.0},"x09":{"count":25870.0,"min":324.0,"max":1539.0},"x10":{"count":25870.0,"min":0.56,"max":0.82},"x11":{"count":25870.0,"min":2790.0,"max":7855.0},"x12":{"count":25870.0,"min":286.0,"max":1000.0},"x13":{"count":25870.0,"min":500.0,"max":1300.0},"x14":{"count":25870.0,"min":0.0,"max":1290.28},"x15":{"count":25870.0,"min":0.0,"max":0.992523077},"x16":{"count":25870.0,"min":0.0,"max":0.02},"y":{"count":25870.0,"min":11602.1724,"max":1690000.0}}'
data_stats = pd.read_json(StringIO(data_stats_json))
data_stats
```

```python

dflen = 25870

df1 = pd.DataFrame()
```

```python
for c in str_cols:
    df1[c]= pd.Series(rnd.choice(str_cols[c]) for i in range(dflen))
    
for c in data_stats:
    df1[c]= pd.Series(rnd.uniform(data_stats.loc['min', c], data_stats.loc['max', c]) for i in range(dflen))
    
df1
```

```python
df1.to_csv('dataset+target.csv')
```

```python
df1.drop(columns=['y']).to_csv('dataset.csv')
```

```python
import sqlite3 as sq
table_name = "dataset" # table and file name

conn = sq.connect('{}.sqlite'.format(table_name)) # creates file
df1.to_sql(table_name, conn, if_exists='replace', index=False) # writes to file
conn.close()
```
